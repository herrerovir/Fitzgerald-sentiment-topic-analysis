## Sentiment Analysis Summary

The goal of sentiment analysis was to evaluate the emotional tone of each chapter using two tools:

* **VADER**, a rule-based model designed for short texts like social media posts
* **CardiffNLP RoBERTa**, a transformer-based model trained on a broader range of labeled sentiment data

VADER generates both a sentiment score and a categorical label, while Cardiff RoBERTa provides only a categorical label. To allow for a consistent comparison, the transformer-based outputs were mapped to numerical scores. Both models ultimately classified sentiment as **positive**, **neutral**, or **negative**.

VADER performed reasonably well on *The Great Gatsby*, but struggled to capture the nuance and complexity of the prose in the other two novels. In contrast, Cardiff RoBERTa consistently produced more accurate and sensitive results, especially when identifying subtle emotional shifts. It also tended to interpret Fitzgerald’s language with a darker tone, particularly in *This Side of Paradise* and *The Beautiful and Damned*.

## Topic Modeling Summary

The goal of topic modeling was to uncover the dominant themes and patterns within each novel, helping to trace how key ideas develop and shift over time. Topic modeling was conducted using two approaches:

- **Latent Dirichlet Allocation (LDA)**, a probabilistic, unsupervised algorithm that models documents as mixtures of topics and topics as distributions of words. It’s widely used for traditional topic modeling tasks and works best on large, clearly structured corpora.

- **BERTopic**, a transformer-based model that combines BERT embeddings with clustering techniques to group semantically similar documents. It is particularly effective for capturing nuanced and context-rich topics in literary or smaller datasets.

Although LDA was initially tested, it did not perform well on this corpus. The topics it produced were weakly differentiated, with repetitive keywords and high overlap between documents. In contrast, BERTopic generated distinct and coherent topic clusters that aligned more closely with the thematic and emotional flow of Fitzgerald’s writing.

## Emotional and Thematic Progression in Each Novel

This section looks closely at how sentiment and topics develop throughout each novel. By examining changes chapter by chapter, it reveals the emotional and thematic journeys of the characters and how Fitzgerald’s stories unfold over time. Comparing sentiment and topic results helps show where emotional tone and themes align or differ across the novels.

### **This Side of Paradise**

In This Side of Paradise, sentiment and topic progress closely together. Early chapters show neutral sentiment paired with themes like Emotional Reflection and Social and Conflict, capturing Amory’s youth and social ambitions. As the story moves on, sentiment shifts negative in chapters focused on Nature and Atmosphere and Emotional Reflection, reflecting mood changes and growing personal struggles. The last chapters stay negative, with topics like Communication and Work Life and Money and Change, showing Amory’s move toward disillusionment and financial worries.

This clear link between sentiment and topic progression highlights the novel’s shift from inner feelings to external challenges, mirroring Amory’s journey from idealism to disappointment. Negative sentiment often matches topics related to social tension or material problems, reinforcing the sense of crisis.

| Chapter      | Sentiment Label (Cardiff RoBERTa) | Thematic Label (BERTopic)   | 
| ------------ | --------------------------------- | --------------------------- | 
| CHAPTER I    | Neutral                           | Emotional Reflection        | 
| CHAPTER II   | Neutral                           | Social and Conflict         | 
| CHAPTER III  | Negative                          | Nature and Atmosphere       | 
| CHAPTER I    | Negative                          | Emotional Reflection        | 
| CHAPTER II   | Neutral                           | Love and Relationships      | 
| CHAPTER III  | Negative                          | Social and Conflict         | 
| CHAPTER I    | Negative                          | Nature and Atmosphere       | 
| CHAPTER II   | Negative                          | Communication and Work Life | 
| CHAPTER III  | Negative                          | Money and Change            |

```python
%run ../notebooks/setup_path.py
from config import *

import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image


#  This Side of Paradise data
chapters = ["CHAPTER I", "CHAPTER II", "CHAPTER III", "CHAPTER IV", "CHAPTER V", "CHAPTER VI", "CHAPTER VII", "CHAPTER VIII", "CHAPTER IX"]
sentiments = ["neutral", "neutral", "negative", "negative", "neutral", "negative", "negative", "negative", "negative"]
topics = ["Emotional Reflection", "Social and Conflict", "Nature and Atmosphere", "Emotional Reflection", "Love and Relationships", "Social and Conflict",
          "Nature and Atmosphere", "Communication and Work Life", "Money and Change"]

# Map sentiment and topics to numeric values
sentiment_map = {"positive": 1, "neutral": 0, "negative": -1}
sentiment_numeric = [sentiment_map[s] for s in sentiments]

unique_topics = list(dict.fromkeys(topics))  # Preserve order of first appearance
topic_map = {t: i for i, t in enumerate(unique_topics)}
topic_numeric = [topic_map[t] for t in topics]

# Plot setup
fig, ax1 = plt.subplots(figsize = (17, 5))

# Plot sentiment line
line1, = ax1.plot(chapters, sentiment_numeric, color = "#eeca16", marker = "o", label = "Sentiment")

# Sentiment labels
for i, (x, y) in enumerate(zip(chapters, sentiment_numeric)):
    ax1.text(i + 0.08 , y + 0.03, sentiments[i].capitalize(), color = "#9a8417", ha = "left", fontsize = 9)

# Plot topic line on secondary axis
ax2 = ax1.twinx()
line2, = ax2.plot(chapters, topic_numeric, color = "#6a2119", marker = "o", label = "Topic", linestyle = "--")

# Topic labels
for i, (x, y) in enumerate(zip(chapters, topic_numeric)):
    ax2.text(i + 0.14, y + 0.04, topics[i], color = "#6a2119", ha = "left", va = "center", fontsize = 9)

# Configure ticks and axes
ax1.set_yticks([])
ax2.set_ylabel("Topic ID")
ax1.set_xticks(range(len(chapters)))
ax1.set_xticklabels(chapters, ha = "center")
ax1.tick_params(axis = "x", which = "both", bottom = True, top = False, length = 5)
ax2.tick_params(axis = "y", which = "both", right = True, length = 5)

# Add grid
ax1.grid(True, alpha = 0.4)

# Expand upper limit of secondary y-axis to fit labels
ax1.set_ylim(min(sentiment_numeric) - 0.2, max(sentiment_numeric) + 0.2)

# Adjust x-axis to avoid label cutoff
ax1.set_xlim(-0.5, len(chapters) - 0.5 + 0.4)

# Add legend
fig.legend(handles=[line1, line2], loc = "upper center", bbox_to_anchor = (0.5, 1), ncol = 2, fontsize = 10)

# Title
plt.title("This Side of Paradise: Sentiment and Topic Progression", fontsize = 14, pad = 10)

plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save plot
plt.savefig(SENTIMENT_TOPIC_DIR / "this-side-of-paradise-sentiment-topic-progression.png")
```
